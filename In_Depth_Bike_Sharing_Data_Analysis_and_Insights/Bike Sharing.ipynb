## 17.Create a pie chart that showed the distribution of bike shares by season

```python
Sizes = [BikeShare_by_seoson.get_group(0)['Count'].sum(), BikeShare_by_seoson.get_group(1)['Count'].sum(),
         BikeShare_by_seoson.get_group(2)['Count'].sum(), BikeShare_by_seoson.get_group(3)['Count'].sum()]
Sizes
```

```python
labels = ['Spring' , 'Summer' , 'Fall', 'winter']
```

```python
Fig , Axe = plt.subplots(figsize = (14,8))
Axe.pie(Sizes, labels = labels, autopct = '%1.1f%%' , explode = [0.1,0,0.1,0])
Axe.legend(loc = 'best')
plt.show()
```
