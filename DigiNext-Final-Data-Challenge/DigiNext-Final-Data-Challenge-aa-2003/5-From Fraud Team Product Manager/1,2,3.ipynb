```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import os
```

```python
data_path = os.getenv('DATA_PATH')
products = pd.read_csv(f'{data_path}/BaSalam.products.csv')
reviews = pd.read_csv(f'{data_path}/BaSalam.reviews.csv')
labeled_comments = pd.read_csv(f"{data_path}/all_labeled_comments.csv")
map_category = pd.read_csv(f"{data_path}/category_map.csv")
category_mapping = dict(zip(map_category['sub_category'], map_category['main_category']))
products['main_category'] = products['categoryTitle'].map(category_mapping).fillna("سایر")
products[['categoryTitle','main_category']].sample()
```

### 1-کدام کالاها روند فروششون هم توزیع باقی کالاهای مشابهشون نبوده


```python
x = products[products['sales_count_week']>4].reset_index().sort_values('main_category')
x.shape
```

```python
fig = px.violin(x, x='sales_count_week', color='main_category', box=True, points='all', hover_data=['name'])
fig.update_layout(title_text='Distribution of Numeric Column by Category')
fig.show()
```

```python
fig = px.density_contour(x, x='sales_count_week', color='main_category', hover_data=['name'])
fig.update_layout(title_text='Density Contour of Numeric Column by Category')
fig.show()
```

```python
fig = px.box(x, x='sales_count_week', color='main_category', points='all', hover_data=['name'])
fig.update_layout(title_text='Boxplot of Numeric Column by Category')
fig.show()
```

```python
fig = px.histogram(x, x='sales_count_week', color='main_category', barmode='overlay', marginal='box', opacity=0.7, hover_data=['name'])
fig.update_layout(title_text='Density Contour of Numeric Column by Category')
fig.show()
```
