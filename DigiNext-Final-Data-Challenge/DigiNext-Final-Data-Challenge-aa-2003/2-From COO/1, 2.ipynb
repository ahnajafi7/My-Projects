```python
data = pd.merge(labeled_comments, reviews[['_id', 'productId']])
data = data.groupby('productId').filter(lambda x: len(x) >= 5)
```

```python
x = data['productId'].value_counts().reset_index()
x.index = x['productId']
x['bad_delivery'] = 0
y = data[data['delivery']==1]['productId'].value_counts().reset_index()
x.loc[y['productId'], 'bad_delivery'] = y['count'].values / x.loc[y['productId'], 'count']
```

```python
x.index.name = None
z = x[x['bad_delivery'] > 0.3].sort_values(['bad_delivery','count'], ascending=[False, False])
z = pd.merge(z, products[['_id', 'categoryTitle', 'main_category', 'vendor_id']], left_on='productId', right_on='_id')
z
```
